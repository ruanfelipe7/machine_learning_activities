### 3. Apresentação da quantidade de linhas do conjunto de dados.

### 4. Informações de uma coluna numérica do conjunto de dados. 

```python
tempoSong = dataframe['tempo']
print("Coluna Tempo")
print("Média:", tempoSong.mean())
print("Desvio padrão:", tempoSong.std())
print("Mínimo:", tempoSong.min())
print("Máximo:", tempoSong.max())
```
