```python
lines = len(dataframe)

max_train = int(lines * 0.8)

random.seed(98) 
random_numbers = random.sample(range(1, lines+1), max_train) 

mascara = dataframe["Id"].isin(random_numbers)

trainSet = dataframe[mascara.values]
testSet = dataframe[~mascara.values]


print("Conjunto de Treino:" , len(trainSet))
print("Conjunto de Teste:" , len(testSet))
```

### 4. Usando somente o conjunto de treino, mostre as médias dos valores de cada feature para cada classe.


```python
trainSpecies = trainSet.groupby(['Species'])
trainSpecies.mean()
```

### 5. Crie um algoritmo seu (sem usar aprendizado de máquina) para baseado nas médias das features calculadas no item 2.3. ou mesmo usando outras ideias suas, classificar corretamente as amostras do conjunto de teste. Aperfeçoe o seu algoritmo de forma a obter melhores resultados. **Ideia do algoritmo**

#### A ideia do algoritmo criado é calcular a distância quadrática entre o valor da feature e o valor da média obtido no conjunto de teste. Em seguida eles são somados e o resultado da soma é aplicada a uma operação de raiz quadrada, e o resultado é um score que indica o quão longe os valores das features estão das médias encontradas no conjunto de treino. Ao final, a classe que possui o menor score será escolhida, como a espécie da flor.

```python
## Calculando as médias de cada feature do conjunto de treino 
meanSepalLength = trainSpecies['SepalLengthCm'].mean()
meanSepalWidth = trainSpecies['SepalWidthCm'].mean()
meanPetalLength = trainSpecies['PetalLengthCm'].mean()
meanPetalWidth = trainSpecies['PetalWidthCm'].mean()

## Dividindo o conjunto de testes nas features e no target
X_test = testSet.drop('Species', axis=1)
y_test = testSet['Species']

## Realiza-se os cálculos para obter os erros das features em relação a média para cada uma das classes de flores 
X_test['errorMeanSetosa'] = np.sqrt( ( ( (X_test['SepalLengthCm'] - meanSepalLength['Iris-setosa']) ** 2) +
                                          ( (X_test['SepalWidthCm'] - meanSepalWidth['Iris-setosa']) ** 2) +
                                          ( (X_test['PetalLengthCm'] - meanPetalLength['Iris-setosa']) ** 2) +
                                          ( (X_test['PetalWidthCm'] - meanPetalWidth['Iris-setosa']) ** 2) ) )

X_test['errorMeanVersicolor'] = np.sqrt( ( ( (X_test['SepalLengthCm'] - meanSepalLength['Iris-versicolor']) ** 2) +
                                          ( (X_test['SepalWidthCm'] - meanSepalWidth['Iris-versicolor']) ** 2) +
                                          ( (X_test['PetalLengthCm'] - meanPetalLength['Iris-versicolor']) ** 2) +
                                          ( (X_test['PetalWidthCm'] - meanPetalWidth['Iris-versicolor']) ** 2) ) )

X_test['errorMeanVirginica'] = np.sqrt( ( ( (X_test['SepalLengthCm'] - meanSepalLength['Iris-virginica']) ** 2) +
                                          ( (X_test['SepalWidthCm'] - meanSepalWidth['Iris-virginica']) ** 2) +
                                          ( (X_test['PetalLengthCm'] - meanPetalLength['Iris-virginica']) ** 2) +
                                          ( (X_test['PetalWidthCm'] - meanPetalWidth['Iris-virginica']) ** 2) ) )
    
## Formam-se as três condições, escolhendo os menores valores de cada score
isIrisSetosa = ((X_test['errorMeanSetosa'] <= X_test['errorMeanVersicolor']) & (X_test['errorMeanSetosa'] <= X_test['errorMeanVirginica']))
isIrisVersicolor = ((X_test['errorMeanVersicolor'] <= X_test['errorMeanSetosa']) & (X_test['errorMeanVersicolor'] <= X_test['errorMeanVirginica']))
isIrisVirginica = ((X_test['errorMeanVirginica'] <= X_test['errorMeanSetosa']) & (X_test['errorMeanVirginica'] <= X_test['errorMeanVersicolor']))    

## Aplicam-se as confições estabelecidas, para que o novo campo 'classification' de X_test seja indicado com a espécie de flor cujo o score seja menor
y_pred = np.where(isIrisSetosa, "Iris-setosa", 
         np.where(isIrisVersicolor, "Iris-versicolor",
         np.where(isIrisVirginica, "Iris-virginica", 0)))

## Calcula-se a quantidade de acertos, comparando-se a escolha do algoritmo com os valores de targets corretos, contidos no vetor y_test
numberHits = np.sum(y_pred == y_test)
numberTests = len(y_test)

## Calcula-se a acurácia do algoritmo
accuracy = numberHits / numberTests
print("Porcentagem de acertos: ", accuracy * 100, "%")
```
