```python
import pandas as pd
import numpy as np
```

#### 1. Resolver a classificação do Iris dataset para k=3 e k=5 com essas funções que implementamos no seguinte documento colab. Usando 70% dos dados para treino e 30% para teste. Ou seja: não é para usar a implementação do scikit para o kNN, nem o split_train_test do scikit.


```python
def euclidean_distances(X, X_row):
    X_ = (X - X_row) ** 2
    return np.sum(X_, axis=1) ** 0.5
```

```python
def get_neighbors(X_train, test_row, k):
    distances = euclidean_distances(X_train, test_row)
    idx_sort = np.argsort(distances)
    return idx_sort[:k]
```

```python
def predict_classification(X, y, test_row, k):
    idx_sort = get_neighbors(X, test_row, k)
    output_values = y[idx_sort]
    counts = np.unique(output_values, return_counts=True)
    idx_max = np.argmax(counts[1])
    prediction = counts[0][idx_max]
    #print('idx_sort:{}, output_values:{}, prediction:{}'.format(idx_sort, output_values, prediction))
    return prediction
```

```python
dataframe = pd.read_csv('/content/drive/MyDrive/ufc - ec/Aprend_Maquina/datasets/Iris.csv', index_col=0)
#dataframe = pd.read_csv('Iris.csv', index_col=0)
df = dataframe.sample(frac=1, random_state=21)
lenTrainSet = int(df.shape[0]*0.7)
trainSet = df[:lenTrainSet]
testSet = df[lenTrainSet:] 
```

```python
trainSet
```

```python
testSet
```

```python
X_train = trainSet.drop('Species', axis=1).values
y_train = trainSet['Species'].values
X_test = testSet.drop('Species', axis=1).values
y_test = testSet['Species'].values
```

```python
def knn_predict(X_train, y_train, X_test, k):
  y_prediction = []
  for x in X_test :
      prediction = predict_classification(X_train, y_train, x, k)
      y_prediction.append(prediction)
      
  return y_prediction
```

```python
def calculateAccuracy(y_pred, y_test):
    numberHits = np.sum(y_pred == y_test)
    numberTests = len(y_test)
    accuracy = numberHits / numberTests
    print('Porcentagem de acertos: {:.2f}%'.format(accuracy*100))
```

## K = 3

```python
k = 3
y_pred = knn_predict(X_train, y_train, X_test, k)
y_pred
```

```python
calculateAccuracy(y_pred, y_test)
```

## K = 5

```python
k = 5

y_pred = knn_predict(X_train, y_train, X_test, k)
y_pred
```

```python
calculateAccuracy(y_pred, y_test)
```
